```python
import pandas as pd
import numpy as np
from scipy.stats import zscore

df = pd.read_excel("Dataset for Data Analytics (1).xlsx")

print("Dataset Shape:", df.shape)
print(df.head())

print("\nMissing Values:")
print(df.isnull().sum())

numeric_cols = ['Quantity', 'UnitPrice', 'ItemsInCart', 'TotalPrice']

for col in numeric_cols:
    df[col].fillna(df[col].median(), inplace=True)

categorical_cols = [
    'Product','ShippingAddress',
    'PaymentMethod','OrderStatus',
    'CouponCode','ReferralSource'
]

for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

print("\nDuplicate Rows:", df.duplicated().sum())
df.drop_duplicates(inplace=True)

for col in numeric_cols:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = np.where(
        df[col] < lower,
        lower,
        np.where(df[col] > upper,
                 upper,
                 df[col])
    )

df['Date'] = pd.to_datetime(df['Date'])

# Feature 1
df['OrderMonth'] = df['Date'].dt.month

# Feature 2
df['OrderYear'] = df['Date'].dt.year

# Feature 3
df['AverageItemPrice'] = df['TotalPrice'] / df['Quantity']

# Feature 4
df['CartValue'] = df['ItemsInCart'] * df['UnitPrice']

# Feature 5
df['CouponUsed'] = np.where(
    df['CouponCode'].notna(),
    1,
    0
)

# ---------------------------------
# Basic Statistics
# ---------------------------------
print("\nStatistics:")
print(df.describe())

# ---------------------------------
# Correlation
# ---------------------------------
print("\nCorrelation Matrix:")
print(df[['Quantity',
          'UnitPrice',
          'ItemsInCart',
          'TotalPrice',
          'AverageItemPrice',
          'CartValue']].corr())

# ---------------------------------
# Top Products
# ---------------------------------
print("\nTop Products:")
print(df['Product'].value_counts().head(10))

# ---------------------------------
# Payment Methods
# ---------------------------------
print("\nPayment Methods:")
print(df['PaymentMethod'].value_counts())

# ---------------------------------
# Order Status
# ---------------------------------
print("\nOrder Status:")
print(df['OrderStatus'].value_counts())

# ---------------------------------
# Save Clean Dataset
# ---------------------------------
df.to_excel(
    "Project1_Cleaned_Dataset.xlsx",
    index=False
)

print("\nCleaned Dataset Saved Successfully!")
```
